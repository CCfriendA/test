```python
#下载fits文档函数
def downloadPic(url,name,i,err):     
    try:
        pic = requests.get(url, timeout=10)
        print('正在下载第'+str(i)+"张名为"+name+"图")
    except requests.exceptions.ConnectionError:
        print('【错误】当前图片无法下载')
        err.append(i)
    except requests.exceptions.Timeout:
        print('时间超时')
        try:
            pic = requests.get(url, timeout=100)
        except requests.exceptions.Timeout:
            print("再次时间超时,第"+str(i)+"记录")
            err.append(i)
    else:
        fp = open(name, 'wb')
        fp.write(pic.content)
        fp.close()
    finally:
        return 
```
